```python
from text.similarity.USE_Transformer.src.USE_Transformer import USE_Transformer_Similarity
from decimal import Decimal
```

```python

instance = USE_Transformer_Similarity()
instance.read_dataset('../data/SemEval2014_image.csv')
```

```python
instance.load_model('../model/moduleA')
#instance.load_model("https://tfhub.dev/google/universal-sentence-encoder-large/3")
```

```python
sim_scores = instance.predict(instance.sentences_1,instance.sentences_2)
```

```python
file = open('../output/output_semEval2014.txt','w')
for score in sim_scores:
    score = str(Decimal(score).quantize(Decimal('0.00')))
    file.write(score)
    file.write('\n')
```
